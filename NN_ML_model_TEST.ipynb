```python
import os
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# load NN_dataset
filepath = "NN_dataset.csv"
NN_data = pd.read_csv(filepath)
NN_data.columns
```

```python
# set Customer_ID as index
NN_data.set_index('Customer_ID',inplace=True)
```

```python
NN_data.dtypes
```

```python
label_encoder = LabelEncoder()

label_encoder.fit(NN_data['Dependents'])
NN_data['Dependents'] = label_encoder.transform(NN_data['Dependents'])

label_encoder.fit(NN_data['Status_Checking_Acc'])
NN_data['Status_Checking_Acc'] = label_encoder.transform(NN_data['Status_Checking_Acc'])

label_encoder.fit(NN_data['Credit_History'])
NN_data['Credit_History'] = label_encoder.transform(NN_data['Credit_History'])

label_encoder.fit(NN_data['Purposre_Credit_Taken'])
NN_data['Purposre_Credit_Taken'] = label_encoder.transform(NN_data['Purposre_Credit_Taken'])

label_encoder.fit(NN_data['Savings_Acc'])
NN_data['Savings_Acc'] = label_encoder.transform(NN_data['Savings_Acc'])

label_encoder.fit(NN_data['Years_At_Present_Employment'])
NN_data['Years_At_Present_Employment'] = label_encoder.transform(NN_data['Years_At_Present_Employment'])

label_encoder.fit(NN_data['Marital_Status_Gender'])
NN_data['Marital_Status_Gender'] = label_encoder.transform(NN_data['Marital_Status_Gender'])

label_encoder.fit(NN_data['Other_Debtors_Guarantors'])
NN_data['Other_Debtors_Guarantors'] = label_encoder.transform(NN_data['Other_Debtors_Guarantors'])

label_encoder.fit(NN_data['Other_Inst_Plans '])
NN_data['Other_Inst_Plans '] = label_encoder.transform(NN_data['Other_Inst_Plans '])

label_encoder.fit(NN_data['Housing'])
NN_data['Housing'] = label_encoder.transform(NN_data['Housing'])

label_encoder.fit(NN_data['Job'])
NN_data['Job'] = label_encoder.transform(NN_data['Job'])

label_encoder.fit(NN_data['Telephone'])
NN_data['Telephone'] = label_encoder.transform(NN_data['Telephone'])

label_encoder.fit(NN_data['Foreign_Worker'])
NN_data['Foreign_Worker'] = label_encoder.transform(NN_data['Foreign_Worker'])

label_encoder.fit(NN_data['Duration_binned'])
NN_data['Duration_binned'] = label_encoder.transform(NN_data['Duration_binned'])

label_encoder.fit(NN_data['Credit_amount_binned'])
NN_data['Credit_amount_binned'] = label_encoder.transform(NN_data['Credit_amount_binned'])

label_encoder.fit(NN_data['Age_binned'])
NN_data['Age_binned'] = label_encoder.transform(NN_data['Age_binned'])
```

```python
data = NN_data.values
X = data[:, 0:17]
y = data[:, 17]
```

```python
X.shape, y.shape
```

```python
# one-hot encode y (y data is already label encoded)
one_hot_y = to_categorical(y)
one_hot_y
```

```python
#split data to train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, one_hot_y, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# create basic single hidden layer model to test (ignore future version warning)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model_TEST = Sequential()
model_TEST.add(Dense(units=51, activation='relu', input_dim=17))
model_TEST.add(Dense(units=17, activation='relu'))
model_TEST.add(Dense(units=2, activation='softmax'))
```

```python
model_TEST.summary()
```

```python
# Compile the model
model_TEST.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
trained = model_TEST.fit(
    X_train,
    y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model_TEST.evaluate(
    X_test, y_test, verbose=2)
print(f"model_TEST - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
